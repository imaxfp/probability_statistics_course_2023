## Visualizing data


#### Line graph

```python
import matplotlib.pyplot as plt

# Data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
asset_prices = [105, 107, 103, 108, 110, 112, 115, 117, 119, 120, 122, 124]

# Plotting the data
plt.figure(figsize=(10,6))  # Setting the size of the figure
plt.plot(months, asset_prices, label='Asset Price', color='green', marker='o', linestyle='-')

# Adding title and labels
plt.title('Asset Price Trend Over the Year')
plt.xlabel('Month')
plt.ylabel('Asset Price in $')

# Displaying the grid
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# Showing the plot
plt.tight_layout()  # Adjusts the plot parameters for a better layout
plt.legend()
plt.show()
```

#### Bar chart

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data
candidates = ['Candidate A', 'Candidate B', 'Candidate C']
media_channels = ['Channel 1', 'Channel 2', 'Channel 3']
coverage = [
    [10, 15, 9],  # Coverage for Candidate A
    [12, 8, 14],  # Coverage for Candidate B
    [11, 13, 12]  # Coverage for Candidate C
]

# Set up bar positions
bar_width = 0.25
index = np.arange(len(candidates))
positions = [index, index + bar_width, index + 2 * bar_width]

# Plotting data
fig, ax = plt.subplots(figsize=(12,7))
colors = ['red', 'blue', 'green']

for idx, channel in enumerate(media_channels):
    ax.bar(positions[idx], coverage[idx], bar_width, color=colors[idx], label=channel)

# Adjusting the plot
ax.set_xlabel('Candidates')
ax.set_ylabel('Coverage Hours')
ax.set_title('Media Coverage of Presidential Candidates by Channel')
ax.set_xticks(index + bar_width)
ax.set_xticklabels(candidates)
ax.legend()

# Displaying the grid
plt.grid(axis='y', linestyle='--', linewidth=0.5)

# Show the plot
plt.tight_layout()
plt.show()

```

```python
import matplotlib.pyplot as plt
import random

# Data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

# Generating random data
max_price = 320000
min_price = 240000

average_price = [random.randint(min_price, max_price) for _ in months]
homes_sold = [60 - (price-min_price)//2000 + random.randint(-5, 5) for price in average_price]  # Negative correlation logic

# Primary axis (left y-axis for number of homes sold)
fig, ax1 = plt.subplots(figsize=(10,6))
ax1.set_xlabel('Month')
ax1.set_ylabel('Homes Sold', color='blue')
ax1.bar(months, homes_sold, color='blue', alpha=0.6, label='Homes Sold')
ax1.tick_params(axis='y', labelcolor='blue')

# Create a secondary y-axis (right y-axis for average price)
ax2 = ax1.twinx()
ax2.set_ylabel('Average Price ($)', color='green')
ax2.plot(months, average_price, color='red', marker='o', linestyle='--', label='Average Price')
ax2.tick_params(axis='y', labelcolor='red')

# Title for the plot
plt.title('Number of Homes Sold and Average Price Over Months (Random Data)')
fig.tight_layout()  # Adjust layout for better appearance
plt.show()
```

## Pie chart

```python
import matplotlib.pyplot as plt

# Data: Nutritional values for 100g of cooked, skinless chicken breast from USDA
labels = ['Proteins', 'Fats', 'Carbohydrates', 'Water & Others']
sizes = [31, 3.6, 0, 65.4]  # The "Water & Others" category includes water, minerals, etc.

# Colors for each section
colors = ['#FF9999', '#66B2FF', '#99FF99', '#FFCC99']

# Exploding the 1st slice (Proteins)
explode = (0.1, 0, 0, 0)  

# Plotting the Pie chart
fig, ax = plt.subplots()
ax.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90, shadow=True)
ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Displaying the pie chart
plt.title('Nutritional Composition of Chicken Breast (per 100g)')
plt.show()
```
